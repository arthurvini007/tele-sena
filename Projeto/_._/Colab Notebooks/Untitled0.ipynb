```python
import csv
from faker import Faker
import random

# Inicializar o gerador de dados fictícios
fake = Faker('pt_BR')

# Função para gerar dados fictícios
def generate_csv_data(num_records):
    data = []

    for _ in range(num_records):
        record = {
            "Nome completo": fake.name(),
            "email": fake.email(),
            "Data de nascimento": fake.date_of_birth(minimum_age=18, maximum_age=80).isoformat(),
            "genero": random.choice(["Masculino", "Feminino", "Outro"]),
            "nome social": fake.first_name_male() if random.choice([True, False]) else fake.first_name_female(),
            "pais": "Brasil",
            "estado": random.choice(["Pernambuco", "São Paulo", "Rio de Janeiro", "Minas Gerais", "Bahia"]),
            "cidade": generate_city(),
            "bairro": fake.city_suffix(),  # Alterado para fake.city_suffix()
            "logradouro": fake.street_name(),
            "numero": random.randint(1, 5000),
            "complemento": fake.address().split(",")[0],  # Usando a primeira parte do endereço como complemento
            "cep": fake.postcode(),
            "telefone": fake.phone_number(),
            "estado civil": random.choice(["Solteiro(a)", "Casado(a)", "Divorciado(a)", "Viúvo(a)"]),
            "profissao": random.choice(["Professor", "Engenheiro", "Médico", "Advogado", "Analista"]),
            "escolaridade": random.choice(["Ensino Fundamental", "Ensino Médio", "Superior", "Pós-graduação"]),
            "data de cadastro": fake.date_this_decade().isoformat()
        }
        data.append(record)

    return data

def generate_city():
    ag_estate_cities = [
        "Caruaru", "Garanhuns", "Santa Cruz do Capibaribe", "Bezerros", "Limoeiro",
        "Pesqueira", "Surubim", "Arcoverde", "São Bento do Una", "Vertentes"
    ]

    other_state_cities = [
        "São Paulo", "Rio de Janeiro", "Belo Horizonte", "Salvador", "Fortaleza",
        "Brasília", "Curitiba", "Recife", "Manaus", "Porto Alegre"
    ]

    state = random.choice(["Pernambuco", "São Paulo", "Rio de Janeiro", "Minas Gerais", "Bahia"])

    if state == "Pernambuco":
        return random.choice(ag_estate_cities)
    else:
        return random.choice(other_state_cities)

# Gerar dados fictícios
num_records = 1000  # Aumente o número de registros conforme necessário
data = generate_csv_data(num_records)

# Definir o nome do arquivo CSV
csv_file = 'dados_usuarios_completos.csv'

# Criar e salvar o arquivo CSV
with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=data[0].keys())
    writer.writeheader()
    writer.writerows(data)

print(f'Arquivo {csv_file} gerado com sucesso.')
```
