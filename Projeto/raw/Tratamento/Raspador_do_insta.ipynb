Exemplo de uso do insta loader para carregar informa√ß√µes de perfis.

```python
post_shortcode = 'C_BV7A9ox-J'

# Carrega o perfil
profile = instaloader.Profile.from_username(L.context, username)

# Encontra a postagem espec√≠fica pelo shortcode
post = None
for p in profile.get_posts():
    if p.shortcode == post_shortcode:
        post = p
        break

if post:
    print(f"Post encontrado: {post.url}")

    # Itera sobre os coment√°rios com atraso entre as requisi√ß√µes
    for comment in post.get_comments():
        print(f"Usu√°rio: {comment.owner.username}")
        print(f"Coment√°rio: {comment.text}")
        print(f"Data: {comment.created_at_utc.strftime('%Y-%m-%d %H:%M:%S')}")
        print(f"Likes: {comment.likes_count}")
        print('-' * 50)

        # Adiciona um atraso de 5 segundos entre cada coment√°rio
        time.sleep(5)

    # Adiciona um atraso ap√≥s a conclus√£o da leitura dos coment√°rios
    print("Conclu√≠do. Aguardando antes de encerrar o script...")
    time.sleep(10)  # Adiciona um atraso final de 10 segundos antes de encerrar o script

else:
    print("Postagem n√£o encontrada!")
```
